**Semantic Kernel**

* Enterprise grade "Open Source" development kit
* Used to build AI Agents
* Integrate with AI models
* Programmig languages supported: C#, Jana and Python

**Why Semantic Kernel?**

* **Extend Functionality**: Chat based APIs can be extended to support Audio and Video
* **Structured Design**: Modular (plug and play) and Observable (telemetry support)
* **Flexible**: Can swap models without rewriting codebase. Faster deployment
