What are SOLID principes?

An acronym for five principles that help make "software designs" understandable, flexoble and maintainable. These are attributed to Robert C. Martin.

* **Single Responsibility Principle (SRP)**: A entity should have only one  responsibility. E.G. implementing database logic and error handling logic as seperate entities.

* **Open/Closed Principle (OCP)**: Classes, Objects, modules, etc. should be open for extension but closed for modification. While we can 'add' new code, care should be taken it does not impact current code.

* **Liskov Substitution Principle (LSP)**: Subtypes must be substitutable for their base types. The 'child' class inherits the functionality of its 'parent' class and can extend the functionality but not impact the 'parent' class's functionality.

* **Interface Segregation Principle (ISP)**: Clients should not be forced to depend on interfaces they do not use. We should avoid the creation of large, general-purpose interfaces; create specific ones.


* **Dependency Inversion Principle (DIP)**: Entities must depend on abstractions (e.g., interfaces) rather than concrete classes or functions.


**What is Dependency Injection (DI)?**

* A design pattern
* It helps implement the "Dependency Inversion Principle" (DIP) principle, which advocates loose coupling and flexibility.
* Widely used in software development

**What is a Dependency Injection (DI) container?**

* A software component
* Provides necessary objects to other parts of the application
* Abstracts the handling of ojects from other parts of the application
* Can handle the lifecycle of such objects - creating or reusing existing instance

**How does DI container work?**

* We provide information on how to create and manage 'dependent' objects
* When the 'dependent' object is required, the DI retrieves the instance
* The DI can decide if it has to create a new instance everytime the 'dependent' object is called or share the existing one

**Why DI container?**

* **Centralized Management**: All 'dependent' objects are management centrally,making it easier to manage them
* **Loose Coupling**:  Classes need not know their dependencies or create them as these are provided through DI
* **Testing**: We can mock these 'dependencies' via mock objects or stubs thereby, helping testing.

In "Microsoft Semantic Kernel", the Kernel is a "Dependency Injection" container, which manages all the services and plugins necessary to run the AI application.
