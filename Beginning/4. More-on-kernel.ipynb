# I

What is the Kernel in "Semantic kernel"?

* The kernel is a "Dependency Injection" container
* It manages all plugins and connectors we provide, to run the AI Agents
* It is used by nearly every component within the Semantic Kernel SDK.

Compoents to be aware of:

As part of building the Kernel, we need to feed 'components' to it. There are two types of compoents.

* **Services**: These include AI Services e.g. "Chat completion" or "Text Generation",  to run our aplication
* **Plugins**: These are containers for one or more functions that can be made available to our application.

Components in depth:

* **AI Service Connectors**: This is an abstraction layer to expose multiple AI Service types from different providers, through a common interface.

Below are the AI Service Modalities*. Modalities are the different types of input and output formats that AI models can handle. 

1. **Chat Completion**: Facilitates conversational interactions with AI models, maintaining chat history for context
2. **Text Generation**: To generate text based on given prompts
3. **Embedding Generation**: Creates numerical representations of text
4. **Text to Image**: Converts text descriptions into visual images
5. **Image to Text**: Extracts textual information from images
6. **Text to Audio**: Converts text into spoken audio
7. **Audio to Text**: Transcribes audio recordings into text

Semaantic Kernel abstrats these modalities through AI Service Connectors*:

* **Vector Store (Memory connectors)**: This component provides an abstraction layer that exposes vector stores from different providers via a common interface. Note that the Kernel does not use any registered vector store automatically

* **Plugins**: Plugins contain one or more functions that can be invoked by AI Service Modalities or as part of templates.

* **Prompt Templates**: Templates help to define and structure the inputs fed to AI Service Modalities.

* **Filters**: They allow us to inject custom logic into the execution flow of functions and prompts. These can be before or after specific events.
