To build an AI agent using Semantic Kernel, we follow a structured workflow, as described below:

* **Import packages**: Import the necessary packages (e.g., chat, connectors, memory, etc.) to create the AI Agent
  
* **Add AI services**: What type of AI service do we want to create? E.G., 'chat completion'and 'text generation'. There are other experimental services given [here](https://learn.microsoft.com/en-us/semantic-kernel/concepts/ai-services/).
    
* **Enterprise components**: We can leverage enterprise level services like 'filters', 'logging', 'metrics'

* **Build the kernel**: In Python, we can leverage the "AI Service" from the object directly.

* **Add memory**: The ability to store and retrieve information from the agents past interactions, to improve its future actions.

* **Add plugins**: These are the components that are used by your AI services and prompt templates to perform work. E.G., plugins to retrieve data from a database.

* **Create kernel arguments**: These provide a way to give inputs to functions and plugins within the Semantic Kernel. They are in the form of a key-value pair.

* **Create prompts**: Input texts that guide the LLMs. We can configure 'Prompt templages' that provide a tructured and reusable way to define its behavior.

* **Planning**: Helps LLMs call approparite functions to achieve the user's request. Auto function calling is the default behavior in Semantic Kernel, but you can also manually invoke functions as well.

* **Invoke**: Executing the Agent with steps configured above.

While not all the above steps are required for every AI Agent, the order in which they are listed is to be followed.

Semantic Kernel currently supports .Net (C#), Java and Python programming languages and has libraries created in each of these languages.
